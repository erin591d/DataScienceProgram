## Install Anaconda

Visit this website and follow the instructions: https://docs.anaconda.com/anaconda/install/mac-os/
1. Download the graphical macOS installer for your version of Python.
2. Double-click the downloaded file and click continue to start the installation.
3. Answer the prompts on the Introduction, Read Me, and License screens.
4. Click the Install button to install Anaconda in your ~/opt directory.
5. On the Destination Select screen, select Install for me only.
6. Click the continue button.
7. To install PyCharm for Anaconda, click on the link to https://www.anaconda.com/pycharm.
8. Verify your installation.

Anaconda is a free, easy-to-install package manager, environment manager, and Python distribution with a collection of 1,500+ open source packages. An IDE, or Integrated Development Environment, enables programmers to consolidate the different aspects of writing a computer program. Anaconda is great to use because it brings many of the tools used in data science and machine learning with just one install, so it's great for having short and simple setup.

A script is simply a text file containing a set of commands and comments. The script can be saved and used later to re-execute the saved commands. The script can also be edited so you can execute a modified version of the commands. A notebook is an R Markdown document with chunks that can be executed independently and interactively, with output visible immediately beneath the input.

This is what the RStudio screen looks like. On the left, there is the Console. That is where all the code is run. Where the > is is where code can be typed or pasted to be run. On the top right, there is the Environment, which tells the user all the variables and datasets currently loaded into RStudio. Below that is the Plots and Help section. When a graph is plotted in RStudio, it will show up in the bottom right. Additionally, when you type ?function in the Console, the R documentation for that function will appear in the bottom right. In the top left, there is a button with a white sheet and a plus. Clicking this will allow you to create a new R Script, Notebook, or Markdown file.

Writing an R Notebook document is no different than writing an R Markdown document. The text and code chunk syntax does not differ between the two documents. The primary difference is in the interactiveness of an R Notebook. Primarily that when executing chunks in an R Markdown document, all the code is sent to the console at once, but in an R Notebook, only one line at a time is sent. This allows execution to stop if a line raises an error. R Markdown and R Notebook focuses on reproducible batch execution, plain text representation, version control, production output and offers the same editor and tools that are used for R scripts. On the other hand, the Jupyter Notebooks focus on outputting inline with code, caching the output across sessions, sharing code and outputting in a single file.

### Tips for Coding in R
- If you are unsure about a function in R, type ?function in the Console, and the R documentation for that function will appear.
- R indexing starts at 1, as opposed to Python which starts at 0
- To assign a value to a variable, use the <- instead of the =
