# Matplotlib Exercises

## Exercise 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

sales = pd.read_csv('company_sales_data.csv')
sales
```

```python
x = sales['month_number']
y = sales['total_profit']

plt.plot(x, y)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

## Exercise 2

```python
x = sales['month_number']
y = sales['total_profit']

plt.plot(x, y, 'o', linestyle='dotted', color='r', markerfacecolor='red', linewidth=3, label="Profit Trend")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend(loc='lower right')
```

## Exercise 3

```python
x = sales['month_number']

colors = ['green', 'red', 'blue', 'yellow', 'purple', 'black']
markers = ['o', '.', ',', 'x', '+', 'v']
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

for i in range(1, 7):
    plt.plot(x, sales[sales.columns[i]], marker=markers[i-1], color=colors[i-1], label=labels[i-1])
    
plt.xlabel("Month Number")
plt.ylabel("Number of Units")
plt.legend()
```

### Exercise 4

```python
x = sales['month_number']
y = sales['toothpaste']

plt.scatter(x, y)
plt.xlabel("Month Number")
plt.ylabel("Toothpaste")
```
